```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('netflix_data.csv')
```

# Create a new Column in the provided dataset. Column name should be num_of_cast. Total Num of Cast in Movie or Tv-Show

```python
def set_num_of_cast(val):
    if pd.isnull(val):
        return 0
    else:
        return len(val.split(", "))
```

```python
data['num_of_cast'] = data['cast'].apply(set_num_of_cast)
data.head(15)
```
